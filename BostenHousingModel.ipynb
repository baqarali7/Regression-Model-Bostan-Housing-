```python
import numpy
import pandas
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasRegressor
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```python
dataframe = pandas.read_csv("housing2.csv", delim_whitespace=True, header=None)
dataset = dataframe.values
# split into input (X) and output (Y) variables
X = dataset[:,0:13]
Y = dataset[:,13]
```

```python
def create_baseline():
    model = Sequential()
    model.add(Dense(13, activation='relu', input_shape=(13,)))
    model.add(Dense(1,))
    model.compile(optimizer='Adam', loss='mse', metrics=['mae'])
    return model
```

```python
seed = 7
numpy.random.seed(seed)
```

```python
estimator = KerasRegressor(build_fn=create_baseline, epochs=100, batch_size=5, verbose=1)
```

```python
kfold = KFold(n_splits=10, random_state=seed)
results = cross_val_score(estimator, X, Y, cv=kfold)
print("Results: %.2f (%.2f) MSE" % (results.mean(), results.std()))
```

```python
def Small():
    model = Sequential()
    model.add(Dense(7, activation='relu', input_shape=(X.shape[13],)))
    model.add(Dense(1,))
    model.compile(optimizer='Adam', loss='mse', metrics=['mae'])
    return model
```

```python
estimator = KerasRegressor(build_fn=Small, epochs=100, batch_size=5, verbose=1)
```

```python
kfold = KFold(n_splits=10, random_state=seed)
results = cross_val_score(estimator, X, Y, cv=kfold)
print("Results: %.2f (%.2f) MSE" % (results.mean(), results.std()))
```

```python
def Large():
    model = Sequential()
    model.add(Dense(13, activation='relu', input_shape=(X.shape[1],)))
    model.add(Dense(7, activation='relu', input_shape=(X.shape[1],)))
    model.add(Dense(1,))
    model.compile(optimizer='Adam', loss='mse', metrics=['mae'])
    return model
```

```python
estimator = KerasRegressor(build_fn=Small, epochs=100, batch_size=5, verbose=1)
kfold = KFold(n_splits=10, random_state=seed)
results = cross_val_score(estimator, X, Y, cv=kfold)
print("Results: %.2f (%.2f) MSE" % (results.mean(), results.std()))
```
